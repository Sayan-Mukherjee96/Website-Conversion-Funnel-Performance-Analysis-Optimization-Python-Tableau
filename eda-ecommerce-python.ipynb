```python
plt.figure(figsize=(5,8))

plt.title("No. of Session By Utm Source")
counts = website_session["utm_source"].value_counts()

plt.bar(counts.index, counts.values)

plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(6,6))

plt.title("No. of Session by Device Type")

counts = website_session["device_type"].value_counts()

plt.bar(counts.index, counts.values)

plt.xticks(rotation=45)
plt.show()
```

```python
merged_df = pd.merge(order, website_session,
                     on="website_session_id",
                     how="left")
```

```python
plt.figure(figsize=(8,6))

sns.barplot(
    data=merged_df,
    x="device_type",
    y="revenue",
    estimator=np.sum
)

plt.title("Total Revenue by Device Type")
plt.xlabel("Device Type")
plt.ylabel("Total Revenue")
plt.show()
```

```python
order["created_at"] = pd.to_datetime(order["created_at"])

order["date"] = order["created_at"].dt.year

yearly_revenue = order.groupby("date")["revenue"].mean().reset_index()

plt.figure(figsize=(6,6))

sns.lineplot(
    data=yearly_revenue,
    x="date",
    y="revenue"
)

plt.title("Yearly Average Revenue")
plt.xlabel("Date")
plt.ylabel("Average Revenue")

plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8,6))

sns.barplot(
    data=merged_df,
    x="utm_source",
    y="revenue",
    estimator=np.sum
)

plt.title("Total Revenue by Utm Source")
plt.xlabel("Utm Source")
plt.ylabel("Total Revenue")
plt.show()
```

```python
website_session["created_at"] = pd.to_datetime(order["created_at"])

website_session["date"] = website_session["created_at"].dt.year

yearly_session = website_session.groupby("date")["website_session_id"].count().reset_index()

plt.figure(figsize=(6,6))

sns.lineplot(
    data=yearly_session,
    x="date",
    y="website_session_id"
)

plt.title("Yearly Total Session")
plt.xlabel("Date")
plt.ylabel("No. of Session")

plt.xticks(rotation=45)
plt.show()
```

```python
order["created_at"] = pd.to_datetime(order["created_at"])

order["year"] = order["created_at"].dt.year  

yearly_order = order.groupby("year")["items_purchased"].mean().reset_index()

plt.figure(figsize=(6,6))

sns.lineplot(
    data=yearly_order,
    x="year",
    y="items_purchased"
)

plt.title("Yearly Average Items Purchased")
plt.xlabel("Year")
plt.ylabel("Average Items Purchased")

plt.xticks(rotation=45)
plt.show()
```
